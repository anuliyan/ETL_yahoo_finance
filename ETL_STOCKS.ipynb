```python
import pandas as pd
import datetime
```

```python
#Extracting data from URL
stocks_url = 'https://finance.yahoo.com/most-active?offset=0&count=50'
tables = pd.read_html(stocks_url)
stock_df = tables[0]
stock_df.head()
```

```python
# Dropping columns 

del stock_df['Avg Vol (3 month)']
del stock_df['PE Ratio (TTM)']
del stock_df['Market Cap']
del stock_df['52 Week Range']
```

```python
stock_df
```

```python
stock_df = stock_df.rename(columns={'Price (Intraday)': 'Price','% Change':'Change'})
```
